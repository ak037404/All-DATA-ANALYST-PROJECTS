```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set theme
sns.set(style="whitegrid", palette="pastel")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Average Tip Across Days and Times
sns.barplot(data=tips, x='day', y='time', color='grey')
plt.title('Average Tip by Day and Time')
plt.show()
```

```python
# Gender-wise Tip Comparison
sns.boxplot(data=tips, x='sex', y='tip', hue='smoker')
plt.title("Tip Comparison by Gender and Smoking Status")
plt.show()
```

```python
# Total Bill vs Tip Relationship
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex", aspect=1.2)
plt.title("Relationship Between Total Bill and Tip")
plt.show()
```

```python
# Party Size vs Tip %
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
sns.barplot(data=tips, x='size', y='tip_pct', errorbar=None, color='purple')
plt.title("Average Tip Percentage by Party Size")
plt.show()
```

```python
# Day with Highest Average Bill
avg_bill = tips.groupby("day")["total_bill"].mean().sort_values(ascending=False)
print(avg_bill)
```

```python
# Distribution Plots
sns.histplot(tips["total_bill"], kde=True)
plt.title("Distribution of Total Bill")
plt.show()

sns.histplot(tips["tip"], kde=True)
plt.title("Distribution of Tips")
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Survival Percentage
survival_rate = titanic['survived'].mean() * 100
print(f"Overall Survival Rate: {survival_rate:.2f}%")
```

```python
# Gender & Class Survival
sns.barplot(data=titanic, x='sex', y='survived', hue='class')
plt.title("Survival by Gender and Passenger Class")
plt.show()
```

```python
# Age Influence
sns.histplot(data=titanic, x='age', hue='survived', kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Fare vs Survival
sns.boxplot(data=titanic, x="survived", y="fare")
plt.title("Fare vs Survival")
plt.show()
```

```python
# Embarkation Port Survival
sns.barplot(data=titanic, x="embarked", y="survived")
plt.title("Survival by Embarkation Port")
plt.show()
```

```python
penguins = sns.load_dataset('penguins')
penguins.dropna(inplace=True)
penguins.head()
```

```python
# Body Mass by Species
sns.boxplot(data=penguins, x="species", y="body_mass_g")
plt.title("Average Body Mass by Penguin Species")
plt.show()
```

```python
# Flipper Length by Island
sns.violinplot(data=penguins, x="island", y="flipper_length_mm", hue="species")
plt.title("Flipper Length by Island and Species")
plt.show()
```

```python
# Correlation Heatmap
sns.heatmap(penguins.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot
sns.pairplot(penguins, hue="species")
plt.show()
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
# Price vs Carat
sns.scatterplot(data=diamonds, x="carat", y="price", hue="cut", alpha=0.6)
plt.title("Diamond Price vs Carat by Cut")
plt.show()
```

```python
# Price by Cut
sns.boxplot(data=diamonds, x="cut", y="price")
plt.title("Price Distribution by Cut")
plt.show()
```

```python
# Correlation Matrix
sns.heatmap(diamonds.corr(numeric_only=True), annot=True, cmap="viridis")
plt.title("Feature Correlation with Price")
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
# Heatmap (Seasonal Pattern)
flights_pivot = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True)
plt.title("Monthly Passenger Heatmap")
plt.show()
```

```python
car = sns.load_dataset('car_crashes')
car.head()
```

```python
# Top Accident States
sns.barplot(data=car.sort_values("total", ascending=False).head(10),
            x="total", y="abbrev")
plt.title("Top 10 States by Total Crashes")
plt.show()
```

```python
# Speeding vs Total Crashes
sns.lmplot(data=car, x="speeding", y="total")
plt.title("Speeding vs Total Crashes")
plt.show()
```

```python
# Correlation Heatmap
sns.heatmap(car.corr(numeric_only=True), annot=True, cmap="magma")
plt.title("Crash Factors Correlation")
plt.show()
```
